# Modelo CN3S

```python
%load_ext autoreload
%autoreload 2

import pandas as pd

from cn3s import CN3S, CN3SParams
```

```python
rain = pd.read_excel("../data/C20500000_CN3S.xls", header=None).iloc[9:, [0,8]]
rain = rain.rename(columns={0: "date", 8: "rain"})
rain["date"] = pd.to_datetime(rain["date"], format="%Y-%m-%d")
rain = rain.dropna()
rain = rain.set_index("date")
rain
```

## CN3S class

```python
rain_list = rain["rain"].to_list()
```

```python
params = CN3SParams()
model = CN3S(params)
```

```python
params
```

```python
model.run(rain_list)
```
